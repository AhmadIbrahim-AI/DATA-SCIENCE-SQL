```python
# import required modules
from datetime import datetime, timedelta
import os
import zipfile
import re
from os.path import basename
import shutil
import ftplib
import pathlib


# Declare the function to return all file paths of the particular directory
def retrieve_file_paths(dirName):

  # setup file paths variable
  filePaths = []

  # Read all directory, subdirectories and file lists
  for root, directories, files in os.walk(dirName):
    for filename in files:
        print(filename)
        test = re.search('.*_(\d{4}[_]\d{2}[_]\d{2})_.*', filename)
        if test:
            var=datetime.now() - timedelta(days=1)
            if test[1]==var.strftime('%Y_%m_%d'):
                filePath = os.path.join(root, filename)
                filePaths.append(filePath)
   

  # return all paths
  return filePaths


# Declare the main function
def main():
  yesterday = datetime.now() - timedelta(days = 1)

# Assign the name of the directory to zip
  dir_name = 'E:\BACKUP'

  # Call the function to retrieve all files and folders of the assigned directory
  filePaths = retrieve_file_paths(dir_name)

  # printing the list of all files to be zipped
  print('The following list of files will be zipped:')
  for fileName in filePaths:
    print(fileName)

  # writing files to a zipfile
  zip_file = zipfile.ZipFile('backup_'+ yesterday.strftime('%Y_%m_%d') +'.zip', 'w')
    
  with zip_file:
    # writing each file one by one
    for file in filePaths:
      zip_file.write(file,basename(file), compress_type=zipfile.ZIP_DEFLATED)
    
    

  FTP_HOST = "192.168.3.57"
  FTP_USER = "sary"
  FTP_PASS = "S@ry20!9"

  # connect to the FTP server
  ftp = ftplib.FTP(FTP_HOST, FTP_USER, FTP_PASS)
  # force UTF-8 encoding
  ftp.encoding = "utf-8"

  # local file name you want to upload
  filename = 'backup_'+ yesterday.strftime('%Y_%m_%d') +'.zip'
  ftp.cwd('BACKUP/Daily Backup I7')
  with open(filename, "rb") as file:
        # use FTP's STOR command to upload the file
        ftp.storbinary(f"STOR {filename}", file)
    
        
     
 
  print('backup_'+ yesterday.strftime('%Y_%m_%d') +'.zip file is created successfully!')

# Call the main function
if __name__ == "__main__":
  main()
```
